```python
from collections import defaultdict
import csv
from dataclasses import dataclass

@dataclass
class Row:
    iso: str
    country: str
    description: str
    price_per_message: float


data = []
iso_to_price = defaultdict(list)

with open("SMSPricing.csv") as fp:
    doc = csv.DictReader(fp)

    for row in doc:
        row = Row(
            iso=row["ISO"],
            country=row["Country"],
            description=row["Description"],
            price_per_message=float(row["Price / msg"]),
        )

        data.append(row)
        iso_to_price[row.iso].append(row.price_per_message + 0.05)

for key, value in iso_to_price.items():
    postfix = ""
    if min(value) != max(value):
        postfix = f"\tmin = {min(value):.3f} max = {max(value):.3f})"
    print(f"{key:<2} = {sum(value) / len(value):.3f}{postfix}")
```
